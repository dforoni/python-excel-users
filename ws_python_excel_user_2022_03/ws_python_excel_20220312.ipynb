# <font color='#F51543'> Estrutura de Dados 

Vamos entender como é a estrutura de dados em **Pandas**.

## <font color='#F51543'> Dataframe 

*DataFrame* é uma estrutura de dados tabular bidimensional mutável em tamanho, e pode armazenar diferentes tipos de dados (int, float, string, date e etc.), com eixos rotulados (linhas e colunas), estrutura parecida com uma planilha de Excel/ Google Sheet.

dataframe.PNG


```python
# criando uma lista de textos, listas são indicadas por colchetes
lista = ['Pyladies SP', 'Deborah', 'Excel', 'Python']
type(lista)
```

**Nota:**

Uma lista é uma sequência de valores, assim como uma *string*.

Em uma *string*, os valores são caracteres; por outro lado, em uma lista, os valores podem ser de qualquer tipo. 

Os valores em lista são chamados de elementos ou às vezes chamamos de itens.

```python
# criando um dataframe com a lista criada anteriormente
dataframe = pd.DataFrame(lista)
dataframe
```

```python
# nomeando a coluna
dataframe = pd.DataFrame(lista, columns=['nomes'])
dataframe
```

## <font color='#F51543'> Series

*Seires* é uma matriz rotulada unidimensional capaz de conter qualquer tipo de dados (inteiros, strings (texto), números decimal, objetos Python, etc.), os rótulos dos eixos são chamados de índice. 

Cada coluna no *dataframe* é uma series, é igual uma coluna no Excel/ Google Sheet.

O índice começa no zero, lembrando que no Excel o índice começa no 1.

serie.PNG




```python
# criando uma serie 
series = pd.Series([1,2,3])
series
```

```python
# criando uma serie com índice texto
series = pd.Series([1, 2, 3, 4, 5], index=["a", "b", "c", "d", "e"])
series
```

```python
# criando uma serie com índice começando no n° 1
series = pd.Series([1, 2, 3, 4, 5], index=[1,2,3,4,5])
series
```

## <font color='#F51543'> Filtros com operadores de comparação

Em alguns momentos precisamos usar operadores de comparação para fazer filtros.


operadoresComparacao.PNG

## <font color='#F51543'> Filtros com operadores lógicos
    
Os operadores lógicos em Python:

| **and** | **&** |

| **or** | **|** |

| **not** | **~** |

```python
# criando os filtros
filtro_valor = df['valor'] > 5000
filtro_tipo = df['tipo'] == 'L - LOCAÇÃO DE BENS IMÓVEIS'
```

```python
# filtro com operadores de comparação e lógicos
df[filtro_valor & filtro_tipo]
```

# <font color='#F51543'> Criar colunas e tabelas a partir dos próprios dados

## <font color='#F51543'> Criar colunas a partir de operações aritméticas entre outras colunas

```python
# criando coluna com a soma de duas variáveis
df['valor_atualizado'] = df['valor'] + 100
df.head(2)
```

```python
# criando coluna com a subtração de duas variáveis
df['valor_atualizado_v2'] = df['valor'] - 100
df.head(2)
```

```python
# criando coluna com a multiplicação de um valor fixo
df['valor_atualizado_v3'] = df['valor'] * 1.1
df.head(2)
```

## <font color='#F51543'> Criar colunas dividindo strings 

```python
# separando a coluna tipo em duas partes
split_tipo = df["tipo"].str.split(" ", n = 1, expand = True)
split_tipo
```

```python
# criando coluna com os as letras 
df['letra'] = split_tipo[0]
df.head(2)
```

## <font color='#F51543'> Excluir colunas e linhas

```python
# excluindo coluna 
df.drop(columns='valor_atualizado_v3', axis=1).head(2)
```

```python
# excluindo colunas 
df.drop(columns=['valor_atualizado', 'valor_atualizado_v2'], axis=1).head(2)
```

```python
# excluindo colunas usando o inplace
df.drop(columns=['valor_atualizado', 'valor_atualizado_v2', 'valor_atualizado_v3'], axis=1, inplace=True)
```

```python
df.head(2)
```

## <font color='#F51543'> Tabela Dinâmica

tabelaDinamica.PNG

```python
# somatória dos dados
df.sum(numeric_only=True)
```

```python
# formatando os valores float
pd.options.display.float_format = '{:,.2f}'.format
```

```python
# value counts: pandas.Series.value_counts
df['tipo'].value_counts()
```

```python
# value counts: normalize: bool, default: False
df['tipo'].value_counts(normalize=True)
```

```python
# agrupando valores de mais colunas
groupby_dados  = df.groupby(by='tipo').count()
groupby_dados
```

```python
# agrupando valores de mais colunas e tirando o index
groupby_dados  = df.groupby(by='tipo', as_index=False).count()
groupby_dados
```

```python
# groupby com a função agg, usando as duas conseguimos definir qual coluna vamos agrupar os dados
groupby_dados  = df.groupby(by='tipo', as_index=False)['valor'].agg(['count', 'mean'])
groupby_dados
```

```python
# groupby com a função agg, usando as duas conseguimos definir qual coluna vamos agrupar os dados
groupby_dados  = df.groupby(by=['tipo', 'ano'], as_index=False)['valor'].agg(['count', 'mean'])
groupby_dados
```

# <font color='#F51543'> Inserir novos dados a partir de outras tabelas

## <font color='#F51543'> Proc V

```python
ano_2020 = df[df['ano'] == 2020]
ano_2021 = df[df['ano'] == 2021]
```

```python
pd.merge(ano_2020, ano_2021, how='left', on=['cpf_cnpj'], suffixes=('_2020', '_2021'))
```

## <font color='#F51543'> Unindo as duas tabelas horizontalmente

```python
tabelas = [ano_2020, ano_2021]
```

```python
df_full = pd.concat(tabelas)
df_full
```

## <font color='#F51543'> Formatação dos dados
